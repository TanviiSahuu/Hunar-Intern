```python
# Step 1: Input the data
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data for years and heights
data = {
    "Year": [1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980],
    "Height (cm)": [169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168, 169.4, 170.9, 171, 173.9, 174.9, 176, 176.9, 177.1, 176.8]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
```

```python
# Step 2: Calculate the mean height
mean_height = df["Height (cm)"].mean()
print(f"Mean Height: {mean_height:.2f} cm")
```

```python
# Step 3: Visualize the data using appropriate charts
plt.figure(figsize=(12, 10))

# Line plot showing the trend over time
plt.subplot(2, 2, 1)
plt.plot(df["Year"], df["Height (cm)"], marker="o", color="blue")
plt.title('Trend of Heights Over Time')
plt.xlabel('Year')
plt.ylabel('Height (cm)')
plt.grid(True)

# Scatter plot showing the distribution
plt.subplot(2, 2, 2)
plt.scatter(df["Year"], df["Height (cm)"], color="green")
plt.title('Scatter Plot of Heights Over Time')
plt.xlabel('Year')
plt.ylabel('Height (cm)')
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
